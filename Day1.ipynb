```python
import mindquantum as mq
from mindquantum.simulator import Simulator
from mindquantum.core.circuit import Circuit
from mindquantum.core.gates import H, RZ, CNOT, Measure
import numpy as np

print('mindquantum:', mq.__version__)
print('numpy:', np.__version__)
```

```python
nq = 4
p = 0

theta = np.random.uniform(low=-np.pi/4, high=np.pi/4, size=[2*nq-1])

cir = Circuit()
for i in range(nq):
    cir += H.on(i)
    cir += RZ(theta[p]).on(i) ; p += 1
for i in range(1, nq):
    cir += CNOT.on(i, i - 1)
    cir += RZ(theta[p]).on(i) ; p += 1
    cir += CNOT.on(i, i - 1)
for i in range(nq):
    cir += Measure(f'q{i}').on(i)

print(cir)
cir.svg()
```

```python
sim = Simulator('mqvector', cir.n_qubits)
sim.apply_circuit(cir)
print(sim)
```

```python
sim.sampling(cir, shots=10000)
```
